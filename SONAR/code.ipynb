```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test =train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression,Perceptron,RidgeClassifier
```

```python
from sklearn.metrics import confusion_matrix,precision_score,ConfusionMatrixDisplay,accuracy_score
```

```python
models = [LogisticRegression(),Perceptron(),RidgeClassifier()]
for model in models:
    model.fit(x_train,y_train)
    model.predict(x_test)
    x_train_Accurary =model.predict(x_train)
    x_test_Accurary =model.predict(x_test)
    print(f"The accurary of {model} model on training data is :",accuracy_score(x_train_Accurary,y_train))
    print(f"The accurary of {model} model on testing data is : ",accuracy_score(x_test_Accurary,y_test))
```

```python
from xgboost import XGBClassifier
xg = XGBClassifier()
xg.fit(x_train,y_train)
pred_train = xg.predict(x_train)
print("The accurary of model on training data is :",accuracy_score(pred_train,y_train))
pred_test = xg.predict(x_test)
print("The accurary of model on test data is :",accuracy_score(pred_test,y_test))
```
